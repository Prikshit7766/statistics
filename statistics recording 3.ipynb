```python
import numpy as np
import pylab as py
import statsmodels.api  as sm  # to check the data is following GD or not


# np.random generates different random numbers
# whenever the code is executed
# note : when you execute the same code
#the graph looks different than shown below

#random data points generated


data_points=np.random.normal(0,1,100)  # creating normal distributed points and basically there are 100 points
#----> standard normal distribution
# loc-> 0 ---> mean
# scal-> 1----> sd
sm.qqplot(data_points,line='45') # apply qqplot
py.show

# X axis-->
# Y axis-->
# if majority of points fall on the straight line then we can say that it is following GD
```

# log normal distribution

```python
# how to create log values  
```

```python
gfg= np.random.lognormal(0,1,100)
print(gfg)
```

```python
sm.qqplot(gfg,line='45') 
py.show
```

```python
sns.displot(gfg,kind="kde")
```

```python
# now we convert lognormal distribution to normal distribution 
# apllying log 
```

```python
np.log(gfg)
```

```python
sns.displot(np.log(gfg),kind="kde")
```

```python
sm.qqplot(np.log(gfg),line='45') 
py.show
```

```python
# now we can say that it is GD
```

# power law

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#generate non -normal data
orignal_data=np.random.exponential(size=1000) # produce exponetial data

fitted_data,filtted_lambda=stats.boxcox(orignal_data)  # we are using boxcox function to produce alpha value
# to determine your height

sns.displot(orignal_data,kind="kde")# ploting origanl data
sns.displot(fitted_data,kind="kde")# ploting fitted data
plt.legend(loc="upper right")
```

```python
sm.qqplot(orignal_data,line='45') 
py.show
```

```python
sm.qqplot(fitted_data,line='45') 
py.show
```

```python
# another way of doing the same thing
import matplotlib.pyplot as plt
import scipy.stats as stats
import pylab
x_m=10
alpha=15
size=100000 # the size of the sample ( no. of samples )
samples= (np.random.pareto(alpha,size)+1)*x_m  # hear t is adding some noise 
stats.probplot(samples,dist="norm",sparams=(15,10),plot=pylab) # --> similar to qqplot
# dist="pareto" ---> compare with pareto ditribution
plt.show
```
